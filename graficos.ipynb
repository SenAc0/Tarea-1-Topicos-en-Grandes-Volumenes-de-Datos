```python
import matplotlib.pyplot as plt
import numpy as np


# Datos
# Tower Sketch
# K31:
# MAE promedio: 9.42787
# MRE promedio: 8.61177

# K21:
# MAE promedio: 7.48186
# MRE promedio: 5.8687

# ////////// 
# PRECISION, RECALL y F1

# K31
# Precision promedio: 0.0623171
# Recall promedio: 0.86
# F1-score promedio: 0.108809

# K21
# Precision promedio: 0.218043
# Recall promedio: 0.98
# F1-score promedio: 0.340666 
# Jorge  13:50
# COUNT SKETCH

# k=21 w=2M d=10

# MAE promedio: 7.29886
# MRE promedio: 5.68258

# k=31 w=2M d=10

# MAE promedio: 9.23248
# MRE promedio: 8.41314 
# Jorge  14:20
# PRECISION, RECALL y F1

# K21
# Precision promedio: 0.196922
# Recall promedio: 0.932859
# F1-score promedio: 0.307549

# K31
# Precision promedio: 0.0569872
# Recall promedio: 0.804153
# F1-score promedio: 0.0989635



sketches = ["Tower K21", "Tower K31", "Count K21", "Count K31"]

mae = [7.48186, 9.42787, 7.29886, 9.23248]
mre = [5.8687, 8.61177, 5.68258, 8.41314]

precision = [0.218043, 0.0623171, 0.196922, 0.0569872]
recall = [0.98, 0.86, 0.932859, 0.804153]
f1 = [0.340666, 0.108809, 0.307549, 0.0989635]

x = np.arange(len(sketches))
width = 0.25


min_mae = min(mae) 
min_mre = min(mre) 







fig, ax = plt.subplots(figsize=(8, 5))

threshold = min_mae
threshold2 = min_mre
ax.axhline(threshold, color="blue", linestyle="--", linewidth=1.5, label=f"Threshold={threshold}") 
ax.axhline(threshold2, color="orange", linestyle="--", linewidth=1.5, label=f"Threshold2={threshold2}") 
# Marcar mnimos 

ax.bar(x - width/2, mae, width, label="MAE")
ax.bar(x + width/2, mre, width, label="MRE")

ax.set_ylabel("Error")
ax.set_title("Comparacin de MAE y MRE entre sketches")
ax.set_xticks(x)
ax.set_xticklabels(sketches, rotation=20)
ax.legend()
plt.tight_layout()
plt.show()


# Grfico Precision, Recall y F1


min_precision = max(precision)
min_recall = max(recall)
min_f1 = max(f1)


fig, ax = plt.subplots(figsize=(8, 5))

threshold = min_precision
threshold2 = min_recall
threshold3 = min_f1

ax.axhline(threshold, color="blue", linestyle="--", linewidth=1.5, label=f"Threshold={threshold}")
ax.axhline(threshold2, color="orange", linestyle="--", linewidth=1.5, label=f"Threshold2={threshold2}")
ax.axhline(threshold3, color="green", linestyle="--", linewidth=1.5, label=f"Threshold3={threshold3}")

ax.bar(x - width, precision, width, label="Precision")
ax.bar(x, recall, width, label="Recall")
ax.bar(x + width, f1, width, label="F1-score")

ax.set_ylabel("Valor promedio")
ax.set_title("Precision, Recall y F1-score por sketch")
ax.set_xticks(x)
ax.set_xticklabels(sketches, rotation=20)
ax.legend()
plt.tight_layout()
plt.show()


labels = ["Precision", "Recall", "F1-score"]
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # cerrar el polgono



```
